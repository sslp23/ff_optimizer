```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import pandas as pd
import numpy as np
```

```python
line_proj = pd.read_csv("projected data/rb_proj.csv")
```

```python
line_proj["FPTS/G_MISC_proj"] = line_proj["FPTS_MISC_proj"]/17
```

```python
line_s = line_proj[line_proj.Season == 2022]
```

```python
line_s = line_s[~line_s.AVG.isna()]
```

```python
line_s = line_s.fillna(0)
```

```python
y_pred = line_s["FPTS/G_MISC_proj"]
y_real = line_s["FPTS/G_MISC"]
```

```python
def evaluate(y_test, y_pred):
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print("mae: ", mae)
    print("mse: ", mse)
    print("r2: ", r2)
    return mae, mse, r2
```

```python
stats_fp = evaluate(y_real, y_pred)
```

```python
mdd = pd.read_csv("projected data/base_rbs.csv")
```

```python
yr_rb = mdd.FPTS_TG
yp_rb = mdd.comp
```

```python
stats_me = evaluate(yr_rb, yp_rb)
```

```python
pd.DataFrame([stats_fp, stats_me], columns=["MAE", "MSE", "R2"], index=["FP PROJ", "MODEL PROJ"]).plot(kind="bar", title="FP vs Model Projections Stats", figsize=(15, 10))
```
