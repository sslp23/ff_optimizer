```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import pandas as pd
import numpy as np
```

```python
line_proj = pd.read_csv("rb_proj.csv")
```

```python
line_proj["FPTS/G_MISC_proj"] = line_proj["FPTS_MISC_proj"]/17
```

```python
line_s = line_proj[line_proj.Season == 2022]
```

```python
line_s = line_s[~line_s.AVG.isna()]
```

```python
line_s = line_s.fillna(0)
```

```python
y_pred = line_s["FPTS/G_MISC_proj"]
y_real = line_s["FPTS/G_MISC"]
```
